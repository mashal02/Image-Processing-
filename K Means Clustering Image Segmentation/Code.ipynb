```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import cv2
import random
labels=[]
np.set_printoptions(threshold=sys.maxsize)

# Read in the image
image = cv2.imread('Image1.bmp')

# Change color to RGB (from BGR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
imgarr = np.array(image)

def getdatapoints(imgarr):
    x,y=np.shape(imgarr)
    datapoints=[]
    for i in range(x):
        for j in range(y):
            if imgarr[i][j]==255:
                arr=[i,j]
                datapoints.append(arr)
   # print(datapoints)            
    return datapoints

def getcentroids(datapoints, K):    
    centroids = []

    #Place K centroids at random locations
    for i in range(K):
        centroid = datapoints[random.randrange(0,len(datapoints))]
        centroids.append(centroid)
   # print("Centroids: ",centroids)
    return centroids


def euclddist(pt1,pt2):
    dist= ((pt1[0]-pt2[0])**2 + (pt1[1]-pt2[1])**2)**0.5
    return dist

def new_centroids(newimgarr, K):
    global labels
    clusterpoints=[]
    for i in range (K):
        arr=[]
        clusterpoints.append(arr)
        
    x,y=np.shape(newimgarr)
    datapoints=[]
    for i in range(x):
        for j in range(y):
            for r in range (K):
                if newimgarr[i][j]==labels[r]:
                   # print(newimgarr[i][j])
                    arr=[i,j]
                    clusterpoints[r].append(arr)
                   # print(clusterpoints[r])

    means=[]
    for cluster in clusterpoints:
        sumx=0 
        sumy=0
        for point in cluster:
            sumx=sumx+point[0]
            sumy=sumy+point[1]
        avgx=sumx/len(cluster)   
        avgy=sumy/len(cluster)
        arr=[int(avgx),int(avgy)]
        means.append(arr)
    return means

def colors(newimg,centroids):
    backtorgb = cv2.cvtColor(newimg,cv2.COLOR_GRAY2RGB)
    x,y=np.shape(newimg)
    for i in range(x):
        for j in range(y):
            if newimg[i][j]==1:
                backtorgb[i][j]=(0,0,255)
            if newimg[i][j]==2:
                backtorgb[i][j]=(0,255,0)
    for centr in centroids:
        m=centr[0]
        n=centr[1]
        backtorgb[m][n]=(0,255,255)
    return backtorgb

def assigncluster(imgarr,k,iterations):
    global labels
    datapoints=getdatapoints(imgarr)
    print(datapoints)
    f=k+1
    for i in range(1,f):
        labels.append(i)
    centroids=getcentroids(datapoints,k)
    print("Randomly initialized centroids: ",centroids)
    x,y=np.shape(imgarr)
    for point in datapoints:
        distances=[]
        for j in range(k):
            distances.append(euclddist(point,centroids[j]))
           # print("Distances: ",distances)
        minm=distances[0]
        index=0
        for dist in range(1,k):
            if distances[dist]<minm:
                minm=distances[dist]
               # print(minm)
                index=dist
                #print("index: ",dist)
        imgarr[point[0]][point[1]]=labels[index]  
    
    c1=colors(imgarr,centroids)
    cv2.imwrite("1.bmp",c1)
    #repeat above steps with n number of iterations but with new centroids
    n=1
    while(n<=iterations):
        #now updating centroids
        newcent=new_centroids(imgarr,k)
        print("Iteration # ",n+1,"Centroids: ",newcent)
        for point in datapoints:
            distances=[]
            for j in range(k):
                #check
                distances.append(euclddist(point,newcent[j]))
            minm=distances[0]
            index=0
            for dist in range(1,k):
                if distances[dist]<minm:
                    minm=distances[dist]
                    index=dist
            imgarr[point[0]][point[1]]=labels[index]
        colored=colors(imgarr,newcent)    
        cv2.imwrite(str(n+1)+".bmp",colored)
        n=n+1
    return imgarr   

   


assigncluster(imgarr,2,4)

                                                                                          
```
